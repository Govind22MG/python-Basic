 6. Explain the concept of generators in Python and how they are defined.

  -> Generators are a special type of function in Python used to create iterators in a simple, memory-efficient way. They allow you to produce a sequence of values one at a time, instead of returning them all at once. This is especially useful for generating large (or infinite) sequences, since you don't need to keep all the results in memory.


 define a generator just like a normal function, but use the yield statement instead of return:

examples:
 Imagine you are at a railway station and there’s a ticket dispenser machine. Each time you press a button, the machine gives you the next ticket number—one at a time. The machine doesn’t print all the tickets at once and hand them to you in a big stack. Instead, it waits for you to ask ("next ticket, please!") and then prints the next ticket.


* Generator Function: Acts like the ticket dispenser that gives one ticket at a time.


* Yield: Each time you press the button, it's as if the machine uses yield to give you the current ticket.

* Lazy/Efficient: The machine never wastes paper printing tickets you never ask for. Only the tickets you request are produced.

def ticket_dispenser():
    ticket = 1
    while True:
        yield ticket      # Like handing out a ticket when asked
        ticket += 1       # Get the next ticket ready

# Get tickets as needed

tickets = ticket_dispenser()

print(next(tickets))  # Output: 1

print(next(tickets))  # Output: 2

print(next(tickets))  # Output: 3

print(next(tickets))  # Output: 4  .....








 7. What are the advantages of using generators over regular functions?

  -> advantages of using generators over regular function are:

*  Memory Efficiency:
Generators produce one value at a time and do not store the entire sequence in memory. Regular functions that return a list create the full list in memory, which can be wasteful or even impossible for large data sets.


* Lazy Evaluation (On-Demand Computation):
With a generator, values are produced only when needed (“lazily”), not all at once. This lets you start processing and producing results right away, and you never create unused data.


*  Improved Performance:
As generators pause and resume their state, their performance is often faster and more scalable when handling large data or complex calculations, since they avoid unnecessary computation and memory steps.

Example :


Suppose you want to read a very large file line by line:

With a generator, you can process each line as it comes, using little memory.


With a normal function returning all lines as a list, the entire file would be loaded into memory at once, which may not be practical for huge files.

 8. What is a lambda function in Python and when is it typically used?

  -> A lambda function in Python is a small, anonymous (nameless) function defined with the lambda keyword, rather than the usual def used for regular functions. Lambda functions can have any number of input arguments but can only contain a single expression—the result of which is automatically returned.
  

  A lambda function in Python is typically used in situations where you need a small, simple, one-time function without formally defining it with def. The most common use cases include:


* Short, simple operations: When the task can be done in a single expression, and defining a full function would be too verbose.

* One-off functions: When you only need the function temporarily, such as passing it once as an argument, without reusing it elsewhere.

* As arguments to built-in functions: Lambda functions are frequently used with functions like map(), filter(), sorted(), and reduce() where you apply a function to elements of a collection.

  ex. add = lambda a, b: a + b

 9. Explain the purpose and usage of the `map()` function in Python.

  -> The map() function in Python is a built-in tool for applying a specified function to every item in one or more iterables (like lists, tuples, or strings) and returning an iterator that produces the results one by one.

Purpose of map()


* Transform Data Efficiently: map() is designed to process and transform each element of an iterable without the need for an explicit for-loop, making code cleaner and more concise.

* Lazy Evaluation: It returns a "map object" (an iterator), so elements are only computed when you need them, which can save memory, especially with large data sets.

* Batch Operations: Easily apply a function to each item in one or more collections at once

ex.  nums = [2, 3, 4, 5]

doubled = map(lambda x: x * 2, nums)

print(list(doubled))        # Output: [4, 6, 8, 10]


 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  ->
*  map(): This function applies a given function to every item in an iterable (like a list or tuple) and returns an iterator of the results. It is used to transform each element in a collection.

*  filter(): This function filters items from an iterable based on a function that returns True or False. It returns an iterator of only those items for which the function returns True. It is used to select certain elements.

* reduce(): This function is used to apply a function cumulatively to the items of an iterable, reducing them to a single value (e.g., sum or product). It works by repeatedly applying the function to pairs of items. You need to import it from the functools module.

from functools import reduce

data = [2, 3, 4, 5]

# map: double each element
mapped = map(lambda x: x * 2, data)
print(list(mapped))  # Output: [4, 6, 8, 10]

# filter: keep elements greater than 20
filtered = filter(lambda x: x > 20, data)
print(list(filtered))  # Output: [4, 5]

# reduce: sum all elements
summed = reduce(lambda x, y: x + y, data)
print(summed)  # Output: 113

```python
#  1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

user_input = input("Enter numbers seperated by commas: ")

numbers_list = list(map(int, user_input.split(',')))

result = sum_even_numbers(numbers_list)

print("Sum of even numbers:", result)
```

```python
#  3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

square_numbers = lambda numbers: list(map(lambda num: num ** 2, numbers))

user_input = input("Enter numbers separated by commas: ")
numbers_list = list(map(int, user_input.split(',')))

squared_list = square_numbers(numbers_list)
print(squared_list)

```

```python
#  5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

fib = FibonacciIterator(9)
for num in fib:
    print(num)
```

```python
#  6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exponent):
    current_exp = 0
    while current_exp <= exponent:
        yield 2 ** current_exp
        current_exp += 1


for value in powers_of_two(8):
    print(value)
```

```python
#  7. Implement a generator function that reads a file line by line and yields each line as a string.

with open('example.txt', 'w') as file:
    file.write("This is line 1.\n")
    file.write("This is line 2.\n")
    file.write("This is line 3.\n")

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')

for line in read_file_line_by_line('example.txt'):
    print(line)

```

```python
#  8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 5), (3, 2), (2, 8), (4, 1)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

```python
#  10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    filtered_chars = filter(lambda ch: ch not in vowels, input_string)
    return ''.join(filtered_chars)

user_input = input("Enter a string: ")
result = remove_vowels(user_input)
print("String after removing vowels:", result)


```
